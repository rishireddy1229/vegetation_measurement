```python
import matplotlib.pyplot as plt

# Model names
models = ['KNN','SVM','Gradient Boosting','Random Forest']

# Accuracies
accuracies = [0.8522789676002197,0.9239428885227897,0.9390444810543658,0.9524986271279516]

# Plotting the line graph
plt.figure(figsize=(6, 4))
plt.plot(models, accuracies, marker='o', linestyle='-', color='b')
plt.title('Accuracy of Different Models for Vegetation Classification')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0.8, 1.0)  # Setting y-axis limit for better visualization
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Model names
models = ['KNN', 'SVM', 'GB', 'RF']

# Accuracies
accuracies = [0.8522789676002197, 0.9239428885227897, 0.9390444810543658, 0.9524986271279516]

# Define colors for each model
colors = ['blue', 'green', 'red', 'purple']

# Plotting the line graph for each model with different color
plt.figure(figsize=(6, 4))
for i in range(len(models)):
    plt.plot([models[i]], [accuracies[i]], marker='o', linestyle='', color=colors[i], label=models[i])

for i in range(len(models) - 1):
    plt.plot([models[i], models[i+1]], [accuracies[i], accuracies[i+1]], linestyle='-', color='black')

plt.title('Accuracy of Different Models for Vegetation Classification')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0.8, 1.0)  # Setting y-axis limit for better visualization
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.legend()  # Show legend
plt.tight_layout()
plt.show()
```
